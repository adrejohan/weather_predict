```python
from datetime import datetime, timedelta
from meteostat import Point, Hourly, Stations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import xgboost as xgb
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.impute import KNNImputer

color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
station_ids = ['47999', '47727','47662']

start = datetime.now() - timedelta(days=365*2)
end = datetime.now() - timedelta(hours=1)
```

```python
data_frames = []
for station in station_ids:
    data = Hourly(station, start, end).fetch()
    if not data.empty:
        data_frames.append(data)
```

```python
if data_frames:
  for df in data_frames:
    df['u'] = df['wspd'] * np.cos(np.deg2rad(df['wdir']))
    df['v'] = df['wspd'] * np.sin(np.deg2rad(df['wdir']))
  combined_data = pd.concat(data_frames).groupby(level=0).agg({
      'temp': 'mean',
      'dwpt': 'mean',
      'rhum': 'mean',
      'prcp': 'mean',
      'u': 'mean',
      'v': 'mean',
      'pres': 'mean',
      'coco': lambda x: x.mode()[0] if not x.mode().empty else np.nan
  })
  combined_data['wspd'] = np.sqrt(combined_data['u']**2 + combined_data['v']**2)
  combined_data['wdir'] = np.rad2deg(np.arctan2(combined_data['v'], combined_data['u'])) % 360

  combined_data.drop(columns=['u', 'v'], inplace=True)

  cols = [col for col in combined_data.columns if col != 'coco'] + ['coco']
  combined_data = combined_data[cols]

  print(combined_data.head())

else:
    print("No data fetched for the specified stations and time period.")
```

```python
tss = TimeSeriesSplit(n_splits=4, test_size=4*24*14*1, gap=24)
fig, axs = plt.subplots(4, 1, figsize=(15, 15), sharex=True)

fold = 0
for train_idx, val_idx in tss.split(combined_data):
    train = combined_data.iloc[train_idx]
    test = combined_data.iloc[val_idx]
    train['temp'].plot(ax=axs[fold],
                          label='Training Set',
                          title=f'Data Train/Test Split Fold {fold}')
    test['temp'].plot(ax=axs[fold],
                         label='Test Set')
    axs[fold].axvline(test.index.min(), color='black', ls='--')
    fold += 1
plt.show()
```

```python
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    df['weekofyear'] = df.index.isocalendar().week
    return df

combined_data = create_features(combined_data)
```

```python
feature_columns = ['temp', 'dwpt', 'rhum', 'prcp', 'wspd', 'wdir', 'pres']

def add_lags(df, combined_data, feature_columns):
    for col in feature_columns:
        target_map = combined_data[col].to_dict()
        for lag in range(1, 25):
            df[f'{col}_lag{lag}'] = (df.index - pd.Timedelta(hours=lag)).map(target_map)
    return df
```

```python
combined_data = add_lags(combined_data, combined_data, feature_columns)
```

```python
tss = TimeSeriesSplit(n_splits=4, test_size=4*24*14, gap=24)
models = {}
predicted_features = pd.DataFrame(index=combined_data.index)
all_preds = {}

for target in feature_columns:
    print(f"Training model for {target}...")

    preds = []
    scores = []

    for train_idx, val_idx in tss.split(combined_data):
        train = combined_data.iloc[train_idx]
        test = combined_data.iloc[val_idx]

        train = create_features(train)
        test = create_features(test)

        FEATURES = [col for col in combined_data.columns if col not in feature_columns + ['coco']]

        X_train = train[FEATURES]
        y_train = train[target]

        X_test = test[FEATURES]
        y_test = test[target]

        reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                               n_estimators=2000,
                               early_stopping_rounds=50,
                               colsample_bytree=0.7,
                               gamma=0.1,
                               objective='reg:squarederror',
                               min_child_weight=1,
                               max_depth=3,
                               learning_rate=0.01)

        reg.fit(X_train, y_train,
                eval_set=[(X_train, y_train), (X_test, y_test)],
                verbose=100)

        y_pred = reg.predict(X_test)
        preds.append((test.index, y_test, y_pred))
        score = np.sqrt(mean_squared_error(y_test, y_pred))
        scores.append(score)

        # Store predictions for coco training
        predicted_features.loc[test.index, target] = y_pred

    models[target] = reg
    all_preds[target] = preds

    print(f'Score across folds for {target}: {np.mean(scores):0.4f}')
    print(f'Fold scores for {target}: {scores}')
```

```python
for target in feature_columns:
    print(f"Plotting comparison for {target}...")
    preds = all_preds[target]
    test_idx, y_test, y_pred = preds[-1]

    plt.figure(figsize=(15, 5))
    plt.plot(test_idx, y_test, label='Validation Data', color='blue')
    plt.plot(test_idx, y_pred, label='Predicted Data', color='red')
    plt.title(f'Comparison of Validation and Predicted Data for {target}')
    plt.legend()
    plt.show()
```

```python
unique_coco_values = sorted(combined_data['coco'].dropna().unique())
coco_mapping = {original: mapped for mapped, original in enumerate(unique_coco_values)}
reverse_coco_mapping = {mapped: original for original, mapped in coco_mapping.items()}

combined_data['coco_mapped'] = combined_data['coco'].map(coco_mapping)
```

```python
print("Training coco classifier...")

coco_preds = []
coco_scores = []

for train_idx, val_idx in tss.split(combined_data):
    train = combined_data.iloc[train_idx]
    test = combined_data.iloc[val_idx]

    train = create_features(train)
    test = create_features(test)

    FEATURES = [col for col in combined_data.columns if col not in ['coco', 'coco_mapped']]

    X_train = train[FEATURES]
    y_train = train['coco_mapped']

    X_test = test[FEATURES]
    y_test = test['coco_mapped']

    clf = xgb.XGBClassifier(base_score=0.5, booster='gbtree',
                            n_estimators=2000,
                            early_stopping_rounds=50,
                            colsample_bytree=0.7,
                            gamma=0.1,
                            objective='multi:softmax',
                            min_child_weight=1,
                            max_depth=3,
                            learning_rate=0.01,
                            num_class=len(coco_mapping))

    clf.fit(X_train, y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            verbose=100)

    y_pred = clf.predict(X_test)
    coco_preds.append((test.index, y_test, y_pred))
    score = accuracy_score(y_test, y_pred)
    coco_scores.append(score)

    # Plot the confusion matrix
    cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[reverse_coco_mapping[i] for i in clf.classes_])
    disp.plot(include_values=True, cmap='viridis', ax=None, xticks_rotation='vertical')
    plt.title('Coco Confusion Matrix')
    plt.show()
```
